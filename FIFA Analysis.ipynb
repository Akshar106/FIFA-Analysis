```python
plt.figure(figsize=(8,5))
plt.bar(df['Home Team Name'].value_counts()[0:5].keys(),df['Home Team Goals'].value_counts()[0:5],color="g")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.bar(df['Home Team Name'].value_counts()[0:5].keys(),df['Away Team Goals'].value_counts()[0:5],color="g")
plt.show()
```

```python
max_goals=df[df['Home Team Goals']>3]
max_goals
```

```python
plt.hist(max_goals['Home Team Goals'])
plt.title("Distribution of Goals")
plt.xlabel("Goals")
plt.ylabel("No. of times")
plt.show()
```

```python
Home_win=df[df['Home Team Goals']>df['Away Team Goals']]
Home_win
```

```python
Away_win=df[df['Away Team Goals']>df['Home Team Goals']]
Away_win
```

```python
plt.figure(figsize=(8,5))
plt.pie(df['Home Team Goals'].value_counts()[0:5],labels=df['Home Team Name'].value_counts()[0:5].keys(),autopct='%0.1f%%')
plt.title("Home Team Goals By Teams")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.pie(df['Away Team Goals'].value_counts()[0:5],labels=df['Away Team Name'].value_counts()[0:5].keys(),autopct='%0.1f%%')
plt.title("Away Team Goals By Teams")
plt.show()
```
